### %matplotlib inline을 설치했기 때문에 사용가능

```python
# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2) # '5.'은 포함하지 않으므로 4.8까지
t
```

```python
# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
data = {'a' : np.arange(50),
        'c' : np.arange(50),
        'd' : np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100 # abs : 절대값
```

```python
data
```

```python
plt.scatter('a', 'b', c='c', s='d', marker='s', data = data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.minorticks_on() # 보조 눈금선
plt.show()
```
