# 네이버 영화 웹 페이지 Crawling(BeautifulSoup)

### 네이버 영화 - 네티즌 평점 전체 페이지 크롤링

```python
import urllib.request
from bs4 import BeautifulSoup
import time
```

```python
url = 'https://movie.naver.com/movie/point/af/list.nhn'
html = urllib.request.urlopen(url)
soup = BeautifulSoup(html, 'html.parser')
print(soup)
```

```python
results = soup.select('td.title')
results
```

```python
for result in results:
    print(list(result.stripped_strings))
```

```python
for result in results:
    print(result.stripped_strings[0])
    # 줄바꿈 문자 \n 출력 X
    output = ''
    for item in result.strings:
        stripped = str(item).strip() # 공백 제거
        # print(stripped)
        if stripped == '':
            continue
        else:
            output += item
    
    # 필요 없는 텍스트는 replace('내용', '')을 사용해서 ''으로 지워준다(데이터 갯수가 적을 경우)
    output = output.replace('신고', '')
    output = output.replace('\n','')
    print(output.strip())
    ''   
```
